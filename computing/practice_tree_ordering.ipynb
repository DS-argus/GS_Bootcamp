# TreeNode definition

```python
class TreeNode():
    def __init__(self, s: str, k: int):
        self.name = s
        self.ary = k
        self.child = [None]*k
```

# Your code for practice (recursive implementation)

```python
class Tree():
    def __init__(self, root: TreeNode) -> None:
        self.root = root
    
    def visit(self, depth: int, Node: TreeNode):
        print("--"*(depth - 1), Node.name)
        
    def __DFT_preorderHelp(self, depth: int, curNode: TreeNode):
        if not curNode:
            return

        self.visit(depth, curNode)

        depth += 1
        for childNode in curNode.child:
            self.__DFT_preorderHelp(depth, childNode)
        
    def DFT_preorder(self):
        self.__DFT_preorderHelp(1, self.root)
```
